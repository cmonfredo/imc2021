```python
import numpy as np
import matplotlib.pylab as plt
import librosa.display
import IPython.display
import librosa
```

```python
def subplots(plotPos,title,data,color,ylabel,xlabel):
    plt.subplot(plotPos[0],plotPos[1],plotPos[2])
    plt.title(title)
    plt.plot(data,color)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
```

```python
def amplitudeModulation(carrier,modulatorFreq,modulatorAmp,modIndex):
    timeSeries = np.linspace(0, 1, len(carrier))
    modulator = modulatorAmp * np.cos (2 * np.pi * modulatorFreq * timeSeries)
    product = (1 + modIndex * modulator) * carrier
    
    subplots([3,1,1],'Modulator signal',modulator,'g','Amplitude','Time')
    subplots([3,1,2],'Carrier signal',carrier,'r','Amplitude','Time')
    subplots([3,1,3],'AM signal',product,'purple','Amplitude','Time')
    
    plt.subplots_adjust(hspace=1)
    plt.rc('font', size=15)
    fig = plt.gcf()
    fig.set_size_inches(16, 9)
    
    return product
```

```python
# Amplitude Modulation on a sine wave
timeSeries = np.linspace(0, 1, 44100)
carrier = np.cos(2*np.pi*400*timeSeries)
modSignal2=amplitudeModulation(carrier,40,0.5,1)
IPython.display.Audio(data=modSignal2, rate=44100)
```

```python
# Amplitude Modulation on a full signal
sig , sr = librosa.load('avm.wav')
modSignal=amplitudeModulation(sig,200,0.5,0.25)
IPython.display.Audio(data=modSignal, rate=sr)
```
